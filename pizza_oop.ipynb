```python
class Pizza:
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
    
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost


class Classic(Pizza):
    def __init__(self):
        super().__init__("Classic Pizza", 10.0)


class Margherita(Pizza):
    def __init__(self):
        super().__init__("Margherita Pizza", 12.0)


class TurkishPizza(Pizza):
    def __init__(self):
        super().__init__("Turkish Pizza", 15.0)


class PlainPizza(Pizza):
    def __init__(self):
        super().__init__("Plain Pizza", 8.0)


class Decorator(Pizza):
    def __init__(self, pizza, sauce):
        super().__init__(pizza.get_description() + ", " + sauce.get_description(),
                         pizza.get_cost() + sauce.get_cost())
        self.pizza = pizza
        self.sauce = sauce


class Olives(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza, Pizza("Olives", 2.0))


class Mushrooms(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza, Pizza("Mushrooms", 3.0))


class GoatCheese(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza, Pizza("Goat Cheese", 4.0))


class Meat(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza, Pizza("Meat", 5.0))


class Onions(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza, Pizza("Onions", 1.0))


class Corn(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza, Pizza("Corn", 2.0))


def main():
    menu_file = open("Menu.txt", "r")
    print(menu_file.read())
    
    pizza_choice = input("Please select a pizza: ")
    if pizza_choice == "1":
        pizza = Classic()
    elif pizza_choice == "2":
        pizza = Margherita()
    elif pizza_choice == "3":
        pizza = TurkishPizza()
    elif pizza_choice == "4":
        pizza = PlainPizza()
    else:
        print("Invalid choice!")
        return
    
    sauce_choice = input("Please select a sauce: ")
    if sauce_choice == "11":
        pizza = Olives(pizza)
    elif sauce_choice == "12":
        pizza = Mushrooms(pizza)
    elif sauce_choice == "13":
        pizza = GoatCheese(pizza)
    elif sauce_choice == "14":
        pizza = Meat(pizza)
    elif sauce_choice == "15":
        pizza = Onions(pizza)
    elif sauce_choice == "16":
        pizza = Corn(pizza)
    else:
        print("Invalid choice!")
        return
    
    total_cost = pizza.get_cost()
    
    name = input("Please enter your name: ")
    tc_id = input("Please enter your TC ID: ")
    cc_num = input("Please enter your credit card number: ")
    cc_cvv = input("Please enter your credit card CVV: ")
    
    print("Thank you for your order, {}!".format(name))
    print("Your order details:")
    print("Pizza: {}".format(pizza.get_description()))
    print("Total Cost: {} TL".format(total_cost))
    print("Customer Information:")
    print("Name: {}".format(name))
    print("TC ID: {}".format(tc_id))
    print("Credit Card Number: {}".format(cc_num))
    print("Credit Card CVV: {}".format(cc_cvv))


    order = {
    "Name": name,
    "TC No": tc_id,
    "Card No": cc_num,
    "Card CVV": cc_cvv,
    "Order Description": pizza.get_description(),
    "Cost": total_cost
    }

    with open('Orders_Database.csv', 'a', newline='') as f:
      fieldnames = ["Name", "TC No", "Card No", "Card CVV", "Order Description", "Cost" ]
      writer = csv.DictWriter(f, fieldnames=fieldnames)

      if f.tell() == 0:
          writer.writeheader()

      writer.writerow(order)

    print("Siparişiniz başarıyla kaydedildi, teşekkür ederiz!")
```

```python
main()
```
